```python
import pandas as pd
from scipy import stats

# Load the dataset from the provided file
# This assumes the file is named 'patients.csv'
try:
    df = pd.read_csv('patients.csv')
except FileNotFoundError:
    print("Error: 'patients.csv' file not found.")
    exit()

# --- T-Test Example: Emergency vs. Surgery Satisfaction ---

# 1. Create two separate groups based on the 'service' column.
# We select the 'satisfaction' column for each of these groups.
emergency_satisfaction = df[df['service'] == 'emergency']['satisfaction']
surgery_satisfaction = df[df['service'] == 'surgery']['satisfaction']

# 2. Perform the independent two-sample t-test.
# We set `equal_var=False` to perform a Welch's T-test, 
# which is more robust and doesn't assume the two groups have equal variance.
# .dropna() is used to ensure missing values don't cause an error.
t_statistic, p_value = stats.ttest_ind(
    emergency_satisfaction.dropna(), 
    surgery_satisfaction.dropna(), 
    equal_var=False
)

# 3. Print the results in a clear format
print("--- T-Test Results: Satisfaction (Emergency vs. Surgery) ---")
print(f"Emergency Mean Satisfaction: {emergency_satisfaction.mean():.2f}")
print(f"Surgery Mean Satisfaction:   {surgery_satisfaction.mean():.2f}")
print("-----------------------------------------------------------")
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value:     {p_value:.4f}")

# 4. Interpret the result
alpha = 0.05
if p_value < alpha:
    print(f"\nConclusion: The difference is statistically significant (p < {alpha}).")
else:
    print(f"\nConclusion: The difference is NOT statistically significant (p >= {alpha}).")
```
